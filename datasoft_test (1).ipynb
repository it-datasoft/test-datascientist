```python
from google.colab import files

uploaded = files.upload()
image_path = list(uploaded.keys())[0]  # Get the uploaded image file name
```

```python
import pytesseract
from PIL import Image
import json
import re
import matplotlib.pyplot as plt
```

```python
# Extract text from image function
def extract_text_from_image(image_path):
    # Open the image using PIL
    image = Image.open(image_path)
    # Use Tesseract to do OCR on the image
    extracted_text = pytesseract.image_to_string(image, lang='fra')  # Specify 'fra' for French
    return extracted_text

# Function to map extracted text to JSON structure
def map_to_json(extracted_text):
    # Initialize the JSON structure
    result = {
        "informationsEpoux": {
            "nom": "",
            "prenom": "",
            "profession": "",
            "dateNaissance": "",
            "lieuNaissance": "",
            "parents": {
                "pere": "",
                "mere": ""
            },
            "temoins": [
                {
                    "nom": "",
                    "prenom": "",
                    "profession": "",
                    "adresse": {
                        "quartier": "",
                        "ville": "",
                        "pays": "",
                        "continent": ""
                    }
                }
            ]
        },
        "informationsEpouse": {
            "nom": "",
            "prenom": "",
            "profession": "",
            "dateNaissance": "",
            "lieuNaissance": "",
            "parents": {
                "pere": "",
                "mere": ""
            },
            "temoins": [
                {
                    "nom": "",
                    "prenom": "",
                    "profession": "",
                    "adresse": {
                        "quartier": "",
                        "ville": "",
                        "pays": "",
                        "continent": ""
                    }
                }
            ]
        },
        "documentType": "certificatMariage",
        "documentNumero": "",
        "dateMariage": "",
        "regimeMariage": "",
        "officierEtatCivil": "",
        "adresseMariage": {
            "quartier": "",
            "ville": "",
            "pays": "",
            "continent": ""
        }
    }

    # Split text into lines for easier processing
    lines = extracted_text.split('\n')

    # Use regex patterns to find specific pieces of information
    for line in lines:
        line = line.strip()
        if re.match(r'Nom de l\'époux:?', line, re.I):
            result["informationsEpoux"]["nom"] = line.split(':')[-1].strip()
        elif re.match(r'Prénom de l\'époux:?', line, re.I):
            result["informationsEpoux"]["prenom"] = line.split(':')[-1].strip()
        elif re.match(r'Profession de l\'époux:?', line, re.I):
            result["informationsEpoux"]["profession"] = line.split(':')[-1].strip()
        elif re.match(r'Date de naissance de l\'époux:?', line, re.I):
            result["informationsEpoux"]["dateNaissance"] = line.split(':')[-1].strip()
        elif re.match(r'Lieu de naissance de l\'époux:?', line, re.I):
            result["informationsEpoux"]["lieuNaissance"] = line.split(':')[-1].strip()
        elif re.match(r'Nom du père de l\'époux:?', line, re.I):
            result["informationsEpoux"]["parents"]["pere"] = line.split(':')[-1].strip()
        elif re.match(r'Nom de la mère de l\'époux:?', line, re.I):
            result["informationsEpoux"]["parents"]["mere"] = line.split(':')[-1].strip()
        elif re.match(r'Nom de l\'épouse:?', line, re.I):
            result["informationsEpouse"]["nom"] = line.split(':')[-1].strip()
        elif re.match(r'Prénom de l\'épouse:?', line, re.I):
            result["informationsEpouse"]["prenom"] = line.split(':')[-1].strip()
        elif re.match(r'Profession de l\'épouse:?', line, re.I):
            result["informationsEpouse"]["profession"] = line.split(':')[-1].strip()
        elif re.match(r'Date de naissance de l\'épouse:?', line, re.I):
            result["informationsEpouse"]["dateNaissance"] = line.split(':')[-1].strip()
        elif re.match(r'Lieu de naissance de l\'épouse:?', line, re.I):
            result["informationsEpouse"]["lieuNaissance"] = line.split(':')[-1].strip()
        elif re.match(r'Nom du père de l\'épouse:?', line, re.I):
            result["informationsEpouse"]["parents"]["pere"] = line.split(':')[-1].strip()
        elif re.match(r'Nom de la mère de l\'épouse:?', line, re.I):
            result["informationsEpouse"]["parents"]["mere"] = line.split(':')[-1].strip()
        elif re.match(r'Document type:?', line, re.I):
            result["documentType"] = line.split(':')[-1].strip()
        elif re.match(r'Numéro de document:?', line, re.I):
            result["documentNumero"] = line.split(':')[-1].strip()
        elif re.match(r'Date de mariage:?', line, re.I):
            result["dateMariage"] = line.split(':')[-1].strip()
        elif re.match(r'Régime de mariage:?', line, re.I):
            result["regimeMariage"] = line.split(':')[-1].strip()
        elif re.match(r'Officier d\'état civil:?', line, re.I):
            result["officierEtatCivil"] = line.split(':')[-1].strip()
        elif re.match(r'Adresse de mariage, quartier:?', line, re.I):
            result["adresseMariage"]["quartier"] = line.split(':')[-1].strip()
        elif re.match(r'Adresse de mariage, ville:?', line, re.I):
            result["adresseMariage"]["ville"] = line.split(':')[-1].strip()
        elif re.match(r'Adresse de mariage, pays:?', line, re.I):
            result["adresseMariage"]["pays"] = line.split(':')[-1].strip()
        elif re.match(r'Adresse de mariage, continent:?', line, re.I):
            result["adresseMariage"]["continent"] = line.split(':')[-1].strip()

    return result

# Main execution
if __name__ == "__main__":
    # Extract text from the uploaded image
    extracted_text = extract_text_from_image(image_path)
    print("Extracted Text:")
    print(extracted_text)

    # Map extracted text to JSON structure
    json_result = map_to_json(extracted_text)

    # Print the final result as JSON
    print(json.dumps(json_result, ensure_ascii=False, indent=4))
```
